## 本課題で利用するデータ
本分析で利用したデータは、`../data/apparel.xlsx`に格納されています。<br>
データの記録期間は、**1990年1月5日**~**1993/12/31**です。

### データ形式

| ds         | Weather | Bargain | y      |
|------------|---------|---------|--------|
| 1/5/1990   | 0       | 0       | 58,507 |
| 1/6/1990   | 0       | 0       | 49,088 |
| 1/7/1990   | 0       | 0       | 50,066 |
| 1/8/1990   | 0       | 0       | 45,954 |

### カラム説明

| カラム名 | 説明 |
|----------|------|
| **ds** | 日付（売上日） |
| **Weather** | 天候フラグ（0=晴れ, 1=雨） |
| **Bargain** | バーゲン施策フラグ（0=通常, 1=バーゲン実施） |
| **y** | 売上金額（円） |

<br>

> **注意**: 休業日のデータは欠損値として記録されています。<br>

## 本課題の目的
本課題では、売り上げデータを用いて売り上げ増加のための施策を提案することである。売り上げ増加のための施策提案に必要なことは、以下の通りである。

### データから知る必要があること
- 長期的・短期的トレンド<br>
    全体的に、増加傾向にあるのか/減少傾向にあるのか
- 季節性, 曜日パターン, イベント(バーゲンセール日)による効果<br>
    イベント効果が高ければ、イベント実施が売り上げ増加のための良施策と評価できる
- 売上の安定度(予測精度)<br>
    売り上げが安定的であれば、品質向上等による全体の底上げが良施策と評価できる
- ノイズの強さ<br>
    ノイズが大きい場合、仕入れコストや過剰在庫などのリスクを持つため、ノイズを小さくする必要がある

### 分析方針
上記の目的をもとに、分析方針を定める。目的として、売上の安定度(予測精度)を確認するため、データすべてを学習データとして利用するのではなく以下の通り学習データとテストデータを分けるモデル1とすべてをTRAINデータとするモデル2の二つを作成することとする。

#### モデルAのデータ分割
| データ区分 | 期間                       |
|:----------|:---------------------------|
| TRAIN     | 1990/01/05 ～ 1992/12/31   |
| TEST      | 1993/01/01 ～ 1993/12/31   |

#### モデルBのデータ分割
| データ区分 | 期間                       |
|:----------|:---------------------------|
| TRAIN     | 1990/01/05 ～ 1993/12/31   |

### モデル設計方針
本分析では、状態空間モデルの観測方程式を以下のように設計する。<br>

$\text{売上}_t= \text{トレンド}_t+ \text{季節要因}_t+ \beta_1 \cdot \text{祝日}_t+ \beta_2 \cdot \text{バーゲン}_t+ \beta_3 \cdot \text{休業日}_t+ \varepsilon_t$

- トレンド<br>
    長期的な売上の上昇・下降を表す学習データの対象期間で成長しているのか衰退しているのかを定量化する。
- 季節要因<br>
    月別や週別の周期的な変動を表すかを確認する。
- 祝日, バーゲン, 休業日<br>
    集客増加が見込まれる場合や売り上げゼロに影響する場合を考慮するため、モデルに反映する
